```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv'
penguins = pd.read_csv(url)
penguins
```

```python
#cheack nan rows
nan_rows = penguins.isna().any(axis=1)
penguins[nan_rows]
```

```python
# we can drop rows where(bill_length_mm,	bill_depth_mm,	flipper_length_mm,	body_mass_g	) is NaN
#it isn't posible to fild
penguins = penguins.dropna(subset=['bill_length_mm'])
```

```python
# can fill van values in sex column based on average body_mass_g
penguins[penguins['sex'].isna()]
```

```python
penguins['species'][penguins['sex'].isna()].unique()
```

```python
male_median_weight=  penguins['body_mass_g'][penguins['sex'] == 'male'].groupby(penguins['species']).median()
male_median_weight
```

```python
female_median_weight=penguins['body_mass_g'][penguins['sex'] == 'female'].groupby(penguins['species']).median()
female_median_weight
```

```python
#cheack nan rows
penguins[penguins['sex'].isna()]
```
