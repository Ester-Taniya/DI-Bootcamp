```python
import pandas as pd
df = pd.read_excel('US Superstore data.xls')
df.info()
```

```python
df.head(10)
```

```python
#clean data

data = df.dropna()

data['Order Date'] = pd.to_datetime(data['Order Date'])

data.head(10)
```

```python
import plotly.express as px

data['Order Date'] = pd.to_datetime(data['Order Date'])

fig_line_chart = px.line(data, x='Order Date', y='Sales', title='Sales Trends over the years')
fig_line_chart.show()

fig_map = px.choropleth(data, locations='State', locationmode='USA-states',
                        color='Sales', title='Sales Distribution by State',
                        scope='usa')


fig_map.show()
```

```python
histogram = (
    ggplot(data.nlargest(15, 'Sales'), aes(x='Product ID', y='Sales', fill='Product ID')) +
    geom_bar(stat='identity') +
    labs(title='Top 10 Products by Sales', x='Product ID', y='Sales') +
    theme_minimal()
)

print(histogram)

scatter_plot = (
    ggplot(data, aes(x='Profit', y='Discount')) + geom_point() +
    labs(title='Relationship between Profit and Discount', x='Profit', y='Discount'))

print(scatter_plot)
```
