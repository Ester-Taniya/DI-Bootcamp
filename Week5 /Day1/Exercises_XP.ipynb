üåü Exercise 2 : Creating A Line Plot For Temperature Variation

Instructions

Create a simple line plot using Matplotlib that represents the temperature variations over a week.
Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]) and label the x-axis as ‚ÄúDay‚Äù and the y-axis as ‚ÄúTemperature (¬∞F)‚Äù.

```python

import matplotlib.pyplot as plt

temp = [72, 74, 76, 80, 82, 78, 75]
week = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

plt.plot(week, temp, marker='o')

plt.xlabel('Day')
plt.ylabel('Temperature (¬∞F)')

plt.title('Temperature Variations Over a Week')

plt.show()
```

üåü Exercise 3 : Visualizing Monthly Sales With A Bar Chart

Instructions

Generate a bar chart using Matplotlib to visualize the monthly sales data for a retail store.


üåü Exercise 4 : Data Visualisation

Instructions

Using the dataset sales_data.csv

Calculate the total quantity of products sold by the company.
Identify the category that had the highest revenue and calculate how much revenue it generated.
Determine the average revenue per sale in the dataset.
Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

```python
import pandas as pd
import matplotlib.pyplot as plt

data= pd.read_csv('sales_data.csv')
data.head(20)
```

```python
#the total quantity of products sold by the company.
gruped_quantity=data['quantity'].groupby(data['product']).count()

print('total quantity of products sold by the company :', gruped_quantity.sum() )
```

```python
# category that had the highest revenue
gd = data['revenue'].groupby(data['category']).sum()

max_product = gd.idxmax()
max_revenue =gd.max()

print(f"The product with the maximum revenue is {max_product} : {max_revenue}.")
```

```python
#the average revenue per sale in the dataset. (if each sale it's a group of rows by date  )
data['revenue'].groupby(data['date']).mean()

# if we need to determine the average revenue for alll dataset
data['revenue'].mean()

```

```python
import matplotlib.pyplot as plt

data['date'] = pd.to_datetime(data['date'])
data['Quarter'] = data['date'].dt.quarter


quarter_labels = {1: 'Q1', 2: 'Q2', 3: 'Q3', 4: 'Q4'}
data['Quarter'] = data['Quarter'].map(quarter_labels)


revenue_by_quarter = data.groupby('Quarter')['revenue'].sum()


plt.bar(revenue_by_quarter.index, revenue_by_quarter)
plt.xlabel('Quarter')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Quarter')
plt.show()
```

Instructions

Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.
Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.


```python
#Plot the function y = x^2 for x values ranging from -10 to 10

x=list(range(-10,11))
y=[i**2 for i in x]
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(title='Plot the function y = x^2 for x values (from -10 to 10)', xlabel='x_values', ylabel='y_values')

plt.show()

```

```python
#bar chart to display the sales data of four different products (A, B, C, D)

import matplotlib.pyplot as plt
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]

plt.bar(products, sales, color= 'green')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales Data of Four Products in a Week')

plt.show()
```

```python
# a pie chart representing the following data about favorite fruits

import matplotlib.pyplot as plt

fruits = ['Apple', 'Bananas', 'Cherries', 'Dates']
procents = [40, 30, 20, 10]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(procents, labels=fruits, autopct='%1.1f%%')
plt.legend(title='Favorite Fruits')

plt.title('Favorite fruits of a group of people')
plt.show()
```
