Instructions

Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.
Sample dataset in code:


data = [12, 15, 13, 12, 18, 20, 22, 21]



```python
import scipy.stats

data = [12, 15, 13, 12, 18, 20, 22, 21]

mean_value = scipy.mean(data)

median_value = scipy.median(data)

variance_value = scipy.var(data)

std_dev_value = scipy.std(data)

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Variance: {variance_value}")
print(f"Standard Deviation: {std_dev_value:}")
```


ðŸŒŸ Exercise 3 : Understanding Distributions

Instructions

Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.



```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean_value = 50
std_dev_value = 10

normal_distribution = norm(loc=mean_value, scale=std_dev_value)

x_values = np.linspace(mean_value - 3 * std_dev_value, mean_value + 3 * std_dev_value, 1000)

plt.plot(x_values, normal_distribution.pdf(x_values), label=f'Mean={mean_value}, Std Dev={std_dev_value}')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density Function (PDF)')
plt.grid(True)
plt.legend()
plt.show()
```

Instructions

Perform a T-test on two sets of randomly generated data. Hint: Use numpy.random.normal() to generate the datasets.



```python
import numpy as np
from scipy.stats import ttest_ind


np.random.seed(0) # random number

data1 = np.random.normal(loc=50, scale=10, size=100) # random numbers
data2 = np.random.normal(loc=51, scale=11, size=100) # random numbers

# Perform T-test
t_statistic, p_value = ttest_ind(data1, data2)

# Print the results
print(f'T-statistic: {t_statistic:.4f}')
print(f'P-value: {p_value:.4f}')

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("The difference between the two datasets is statistically significant. Reject the null hypothesis.")
else:
    print("There is not enough evidence to reject the null hypothesis.")
```
