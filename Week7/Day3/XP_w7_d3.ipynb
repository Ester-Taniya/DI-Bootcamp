ðŸŒŸ Exercise 1: Implementing And Comparing Normalization Techniques
Instructions
Create a synthetic dataset with two features: FeatureA and FeatureB, where FeatureA is in the range of [0, 1000] and FeatureB is in the range of [0, 10].
Apply Min-Max Scaling to this dataset and visualize the results using a scatter plot.
Apply Z-score Normalization to the same dataset and visualize these results as well.
Write a short analysis comparing the effects of both normalization techniques on the dataset. Discuss which technique might be more appropriate based on the distribution and range of the data.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

X = np.random.randint(0, 1000, size=(100, 2))

X_minmax = (X - X.min()) / (X.max() - X.min())

plt.scatter(X_minmax[:, 0], X_minmax[:, 1])
plt.xlabel("FeatureA")
plt.ylabel("FeatureB")
plt.title("Min-Max Scaled Data")
plt.show()

X_zscore = (X - X.mean()) / X.std()

plt.scatter(X_zscore[:, 0], X_zscore[:, 1])
plt.xlabel("FeatureA")
plt.ylabel("FeatureB")
plt.title("Z-score Normalized Data")
plt.show()

print("Min-Max Scaled Data:")
print("Mean:", X_minmax.mean())
print("Standard Deviation:", X_minmax.std())
print()
print("Z-score Normalized Data:")
print("Mean:", X_zscore.mean())
print("Standard Deviation:", X_zscore.std())
```

ðŸŒŸ Exercise 2 : Analyzing The Impact Of Min-Max Scaling On Synthetic Data
Instructions
Create a synthetic dataset with two features: FeatureA and FeatureB. Let FeatureA range from 1 to 100 and FeatureB range from 1,000 to 10,000.
Apply Min-Max Scaling to this dataset using Python and the MinMaxScaler from Scikit-Learn.
Visualize the original and scaled datasets using line plots or scatter plots (using Matplotlib or Seaborn).
Write a brief analysis on how Min-Max Scaling transformed the data, focusing on the change in the range of values for both features.


```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

X = np.random.randint(1, 101, size=(100, 2))
X[:, 1] += 999

scaler = MinMaxScaler()
X_minmax = scaler.fit_transform(X)

plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("FeatureA")
plt.ylabel("FeatureB")
plt.title("Original Data")
plt.show()

plt.scatter(X_minmax[:, 0], X_minmax[:, 1])
plt.xlabel("FeatureA")
plt.ylabel("FeatureB")
plt.title("Min-Max Scaled Data")
plt.show()

print("Original Data:")
print("Mean:", X.mean())
print("Standard Deviation:", X.std())
print()
print("Min-Max Scaled Data:")
print("Mean:", X_minmax.mean())
print("Standard Deviation:", X_minmax.std())
```

ðŸŒŸ Exercise 3: Experimenting With Z-Score Normalization On Simulated Data
Instructions
Generate a simulated dataset with two features: FeatureX and FeatureY. ðŸŒŸ Exercise 4 : Identifying Outliers With Box Plots And Z-Scores
Instructions
Dataset Suggestion: Use the â€œBoston Housing Datasetâ€ available on Kaggle. This dataset contains information about housing in the area of Boston Mass. Itâ€™s a good dataset for practicing outlier detection as it has a variety of features with different scales and distributions.

Load the Boston Housing Dataset into a Python environment.
Use box plots to visually identify potential outliers in various features like â€˜CRIMâ€™ (crime rate), â€˜RMâ€™ (average number of rooms per dwelling), and â€˜MEDVâ€™ (Median value of owner-occupied homes in $1000â€™s).
Calculate Z-scores for these features and identify data points that have a Z-score higher than 3 or lower than -3.
Reflect on how outliers identified by box plots and Z-scores overlap or differ.


```python
import pandas as pd
import seaborn as sns
from scipy.stats import zscore
import matplotlib.pyplot as plt


boston_df = pd.read_csv("HousingData.csv")

selected_features = ['CRIM', 'RM', 'MEDV']

plt.figure(figsize=(12, 6))
for i, feature in enumerate(selected_features, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(x=boston_df[feature])
    plt.title(f'Boxplot of {feature}')

plt.tight_layout()
plt.show()

z_scores = zscore(boston_df[selected_features])

outliers = (np.abs(z_scores) > 3).any(axis=1)

print("Indices of outliers:")
print(boston_df[outliers].index)


overlapping_outliers = boston_df[outliers]
boxplot_outliers = boston_df[
    (boston_df['CRIM'] > boston_df['CRIM'].quantile(0.75) + 1.5 * (boston_df['CRIM'].quantile(0.75) - boston_df['CRIM'].quantile(0.25))) |
    (boston_df['RM'] < boston_df['RM'].quantile(0.25) - 1.5 * (boston_df['RM'].quantile(0.75) - boston_df['RM'].quantile(0.25))) |
    (boston_df['MEDV'] < boston_df['MEDV'].quantile(0.25) - 1.5 * (boston_df['MEDV'].quantile(0.75) - boston_df['MEDV'].quantile(0.25)))
]

print("Overlapping outliers:")
print(overlapping_outliers.index)

print("Outliers identified by box plots:")
print(boxplot_outliers.index)

```

ðŸŒŸ Exercise 5 : Handling Outliers In A Large Dataset
Instructions
Dataset Suggestion: â€œNew York City Airbnb Open Dataâ€ on Kaggle. This dataset provides a comprehensive look at the Airbnb activity in New York City and includes features like price, number of reviews, availability, etc., which are ideal for practicing outlier handling.

Load the New York City Airbnb dataset.
Identify outliers in the â€˜priceâ€™ column using the IQR method.
Apply three different outlier handling techniques: trimming, capping, and transformation.
Analyze the effect of each technique on the distribution of the â€˜priceâ€™ feature.
Discuss the pros and cons of each technique in the context of this dataset.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

airbnb_df = pd.read_csv("AB_NYC_2019.csv")

plt.figure(figsize=(10, 6))
sns.histplot(airbnb_df['price'], bins=50, kde=True)
plt.title('Distribution of Airbnb Prices in NYC')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

Q1 = airbnb_df['price'].quantile(0.25)
Q3 = airbnb_df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (airbnb_df['price'] < lower_bound) | (airbnb_df['price'] > upper_bound)

trimmed_df = airbnb_df[~outliers]

capped_df = airbnb_df.copy()
capped_df['price'] = capped_df['price'].clip(lower=lower_bound, upper=upper_bound)

transformed_df = airbnb_df.copy()
transformed_df['price'] = np.log1p(transformed_df['price'])

plt.figure(figsize=(15, 6))

plt.subplot(1, 4, 1)
sns.histplot(airbnb_df['price'], bins=50, kde=True)
plt.title('Original Distribution')

plt.subplot(1, 4, 2)
sns.histplot(trimmed_df['price'], bins=50, kde=True)
plt.title('Trimmed Distribution')

plt.subplot(1, 4, 3)
sns.histplot(capped_df['price'], bins=50, kde=True)
plt.title('Capped Distribution')

plt.subplot(1, 4, 4)
sns.histplot(transformed_df['price'], bins=50, kde=True)
plt.title('Transformed Distribution')

plt.tight_layout()
plt.show()
```

ðŸŒŸ Exercise 6 : Impact Of Outliers On Data Analysis
Instructions
Dataset Suggestion: â€œWine Quality Datasetâ€ on Kaggle. This dataset contains physicochemical tests of wines and a quality rating, making it suitable for understanding the impact of outliers.

Load the Wine Quality Dataset.
Identify outliers in key features like â€˜fixed acidityâ€™, â€˜residual sugarâ€™, and â€˜alcoholâ€™ using your preferred method.
Calculate the mean and standard deviation of these features with and without outliers.
Analyze how outliers affect the interpretation of these features.
Reflect on the importance of considering outliers in data analysis, especially in a context like wine quality assessment.


```python
import pandas as pd
import matplotlib.pyplot as plt


wine_df = pd.read_csv("WineQT.csv")

features_to_analyze = ['fixed acidity', 'residual sugar', 'alcohol']

plt.figure(figsize=(15, 5))

for i, feature in enumerate(features_to_analyze, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(x=wine_df[feature])
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()

def identify_outliers(data, feature):
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = (data[feature] < lower_bound) | (data[feature] > upper_bound)
    return outliers

outliers_dict = {feature: identify_outliers(wine_df, feature) for feature in features_to_analyze}

summary_data = []

for feature in features_to_analyze:
    without_outliers = wine_df[~outliers_dict[feature]][feature]
    with_outliers = wine_df[feature]

    summary_data.append({
        'Feature': feature,
        'Mean without outliers': without_outliers.mean(),
        'Std without outliers': without_outliers.std(),
        'Mean with outliers': with_outliers.mean(),
        'Std with outliers': with_outliers.std(),
    })

summary_df = pd.DataFrame(summary_data)
print(summary_df)
```
